```python
train_df = pd.read_csv("/kaggle/input/titanic/train.csv", index_col=0)
test_df = pd.read_csv("/kaggle/input/titanic/test.csv", index_col=0)
```

```python
train_df = train_df.drop(["Name", "Ticket", "Cabin"], axis=1)
```

```python
test_df = test_df.drop(["Name", "Ticket", "Cabin"], axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

for item in train_df.columns:
    if item == "Sex" or item == "Embarked":
        le.fit(train_df[item])
        train_df[item] = le.transform(train_df[item])
for item in test_df.columns:
    if item == "Sex" or item == "Embarked":
        le.fit(test_df[item])
        test_df[item] = le.transform(test_df[item])
```

```python
train_df.dropna(inplace=True)
test_df.interpolate(method='linear', inplace=True)
test_df.fillna(method="ffill", inplace=True)
test_df.fillna(method="bfill", inplace=True)
```

```python
X_train = train_df.drop("Survived", axis=1)
y_train = train_df["Survived"]
```

```python
from sklearn.tree import DecisionTreeClassifier

tree_clf = DecisionTreeClassifier(random_state=0)
tree_clf.fit(X_train, y_train)
y_pred = tree_clf.predict(test_df)
print(y_pred)
```

```python
pred_df = pd.DataFrame(y_pred, index=test_df.index, columns=["Survived"])
pred_df
```

```python
pred_df.to_csv("submission.csv")
```
